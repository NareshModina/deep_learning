```python
import os
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn

# Define the input range
x = torch.linspace(-10, 10, 500)

# Define the activation functions
activations = {
    # "Linear": x,
    "Sigmoid": nn.Sigmoid(),
    "Hard_Sigmoid": nn.Hardsigmoid(),
    "Tanh": nn.Tanh(),
    "Rectified Linear Unit (ReLU)": nn.ReLU(),
    "Leaky_relu":nn.LeakyReLU(),
    "ReLU6": nn.ReLU6(),
    "Exponential": lambda x: np.exp(x),
    "Exponential Linear Unit (ELU) (alpha=1.0)": nn.ELU(),
    "Scaled Exponential Linear Unit (SELU)": nn.SELU(),
    "Gaussian Error Linear Unit (GELUE)": nn.GELU(),
    "Sigmoid Linear Unit (SiLU)": nn.SiLU(),
    "Mish": nn.Mish(),
    "Softmax (normalized)": lambda x: np.exp(x) / np.sum(np.exp(x)),
    "Log_Softmax": nn.LogSoftmax(),
    "Softplus": nn.Softplus(),
    "Softsign": nn.Softsign(),
}

# Define colors for each activation function
colors = [
    "blue", "green", "red", "orange", "purple", "cyan", "brown", "pink", "gray", "olive"
]

color = "orange"
# # Create a folder to save the plots
# output_folder = "activation_plots"
# os.makedirs(output_folder, exist_ok=True)

# Generate and save individual plots
for (name, func) in activations.items():
    # For Softmax, use a smaller range to avoid extreme values
    if name == "Softmax (normalized)":
        x_input = np.linspace(-2, 2, 500)
        y = func(x_input)
    else:
        y = func(x)
    
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=name, color=color)
    plt.title(f"A: {name}", fontsize=16)
    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(fontsize=12)
    plt.xlabel("z", fontsize=14)
    plt.ylabel("A(Z)", fontsize=14)
    
#     # Save the plot to the output folder
#     save_path = os.path.join(output_folder, f"{name}.png")
#     plt.savefig(save_path)
#     plt.show()
#     plt.close()

# print(f"All plots with different colors saved in the folder: {output_folder}")
```
